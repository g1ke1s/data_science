```python
bookings['total_kids'] = bookings.children + bookings.babies
bookings.groupby('hotel').total_kids.mean().round(2)
```

```python
bookings['has_kids'] = bookings['total_kids'] >= 1 & True
```

```python
canceled_with_kids = bookings[bookings.has_kids==True][bookings.is_canceled==1].agg({'is_canceled': 'count'})
```

```python
total_with_kids = bookings[bookings.has_kids==True].agg({'is_canceled': 'count'})
```

```python
canceled_with_kids / total_with_kids
```

```python
canceled_without_kids = bookings[bookings.has_kids==False][bookings.is_canceled==1].agg({'is_canceled': 'count'})
```

```python
total_without_kids = bookings[bookings.has_kids==False].agg({'is_canceled': 'count'})
```

```python
(canceled_without_kids / total_without_kids).round(4)
```
