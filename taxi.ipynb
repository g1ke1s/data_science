```python
taxi = pd.read_csv(r'https://stepik.org/media/attachments/lesson/359240/taxi_peru.csv', delimiter=';', parse_dates=['start_at', 'end_at', 'arrived_at'])
taxi.driver_score
```

```python
driver_score_counts = taxi.driver_score.value_counts(normalize=True).mul(100) \
                                                                    .round(2) \
                                                                    .reset_index() \
                                                                    .rename(columns={'driver_score': 'percentage', 'index': 'driver_score'}) \
                                                                    .sort_values('driver_score')
driver_score_counts
```

```python
sns.histplot(data=driver_score_counts, x='driver_score')
```

```python
len('driver_score')
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=driver_score_counts, x='driver_score', y='percentage', color='skyblue')
plt.title('Histogram of Driver Scores')
plt.xlabel('Driver Score')
plt.ylabel('Percentage')
plt.grid(False)
plt.show()
```

```python
rider_score_counts = taxi.rider_score \
                         .value_counts(normalize=True) \
                         .mul(100) \
                         .round(2) \
                         .reset_index() \
                         .rename(columns={'rider_score': 'percentage', 'index': 'rider_score'}) \
                         .sort_values('rider_score')
rider_score_counts
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=rider_score_counts, x='rider_score', y='percentage', color='skyblue')
```
